```python
from sklearn import datasets
```

```python
from sklearn import svm
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris_dataset = datasets.load_iris()
```

```python
def visualise_sepal_data():
    iris = datasets.load_iris()
    X = iris.data[:, :2]
    y = iris.target
    plt.scatter(X[:,0],X[:,1],c=y , cmap=plt.cm.coolwarm)
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.title('Sepal Width & Length')
    plt.show()
visualise_sepal_data()    
```

```python
def visualise_petal_data():
    iris=datasets.load_iris()
    X=iris.data[:,2:]
    y = iris.target
    plt.scatter(X[:,0],X[:,1],c=y,cmap=plt.cm.coolwarm)
    plt.xlabel('petal length')
    plt.ylabel('petal width')
    plt.title('petal width and length')
    plt.show()
visualise_petal_data()
```

```python
iris=datasets.load_iris()
X=iris.data[:,:2]
y=iris.target
C=1.0
```

```python
svc = svm.SVC(kernel='linear', C=C).fit(X, y)
```

```python
lin_svc = svm.LinearSVC(C=C).fit(X, y)
```

```python
rbf_svc = svm.SVC(kernel='rbf', gamma=0.7, C=C).fit(X, y)
```

```python
poly_svc = svm.SVC(kernel='poly', degree=3, C=C).fit(X, y)
```

```python
h=0.2
```

```python
x_min,x_max = X[:,0].min()-1,X[:,0].max()+1
```

```python
y_min,y_max=X[:,1].min()-1,X[:,1].max()+1
```

```python
xx,yy=np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
```

```python
titles=['SVC with linear kernel','LinearSVC(linear kernel)','SVC with RBF kernel','SVC with polynomail']
```

```python
for i, clf in enumerate((svc, lin_svc,rbf_svc,poly_svc)):
    plt.subplot(2,2,i+1)
    plt.subplots_adjust(wspace=0.4,hspace=0.4)
    Z=clf.predict(np.c_[xx.ravel(),yy.ravel()])
    Z=Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm)
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max()) 
    plt.xticks(())
    plt.yticks(())
    plt.title(titles[i])
plt.show()
```
